*   **Pre-training.**

    É uma etapa onde um modelo é treinado em uma grande quantidade de dados não rotulados para aprender os padrões de linguagem. É um processo no qual um modelo entende a estrutura geral da linguagem, como gramáticas, sintaxes e um amplo vocabulário em si.

    Por exemplo: Se você quisesse treinar um modelo de linguagem para entender o português, primeiro coletaria milhões de textos em português, desde livros a artigos, a postagens em redes sociais. Durante esta etapa, o modelo aprende a prever a próxima palavra em uma sequência de palavras. Por exemplo, dada a frase “Océu está”, o modelo aprende que a palavra “azul, fechado, nublado, chuvoso” são algumas opções viáveis nesta sequência.

*   **Transfer Learning.**

    É a prática de aplicar o conhecimento obtido em uma tarefa a uma tarefa relacionada, mas separada. No caso dos LLMs, isso na maioria das vezes envolve ajustar um modelo pré-treinado em trote em um novo conjunto de dados, específico da tarefa.

    Por exemplo: Depois de conclui o pré-treinamento, imagine que deseja que seu modelo de linguagem o ajude a responder perguntas sobre a saúde. Então, você pega um modelo de linguagem pré-treinado e faz uma carga de dados de saúde, contendo perguntas e respostas médicas. O conhecimento geral da língua é transferido e ajustado ao novo trabalho de responder questões de saúde.

*   **Embeddings.**

    São representações vetoriais de palavras, frases ou documentos em um espaço contínuo de alta dimensionalidade. Eles apresentam semelhanças semânticas entre palavras, para que palavras semelhantes sejam colocadas mais próximas no modelo de espaço vetorial.

    Por exemplo: se você imaginar as palavras “rei”, “rainha”, “homem” e “mulher”, então em um espaço de incorporação, as distâncias entre “rei” e “rainha” e “homem” e “mulher” são pequenas, pois estão semanticamente relacionadas uma à outra. Além disso, se fizermos um vetor de operação “rei” - “ homem” + “ mulher”, ele ficará aproximadamente na posição “rainha”.

*   **Transformers.**

    É uma arquitetura de rede neural para processamento de dados de sequência, principalmente base de texto, em oposição à abordagem tradicional em RNNs, onde a sequência é processada nas etapas sucessivas do tempo.

    Por exemplo, no caso de tradução automática se a entrada for "How are you?", o Transformer processa a frase completa e gera a tradução "Como você está?" ao mesmo tempo, considerando o contexto completo da frase.

*   **Attention.**

    É um mecanismo em profunda feedforward em Redes Neurais, que permite que o modelo se concentre em partes mais importantes daquela cadeia de entradas enquanto produz cada elemento da saída.

    Por exemplo, no caso de tradução "O gato está dormindo no sofá" para inglês, o mecanismo de atenção pode focar mais na palavra "gato" quando gera "cat" e na palavra "sofá" quando gera "couch". Isso permite que o modelo capture melhor as relações contextuais entre as palavras.

*   **Fine-Tunning**

    Fine-tuning é a prática de treinar adicionalmente um modelo de linguagem pré-treinado no conjunto de dados de treino da tarefa com a qual estivermos lidando.

    Por exemplo: Depois de pré-treinar um modelo de linguagem em textos gerais, você pode fazer o fine-tuning desse modelo em um conjunto de dados específico de atendimento ao cliente, contendo exemplos de perguntas e respostas de clientes. Isso adapta o modelo para fornecer respostas mais precisas e relevantes no contexto de atendimento ao cliente.

Modelos Seq2Seq combinam codificadores e decodificadores para transformar uma sequência de texto em outra. Exemplos incluem tradução automática e sumarização de texto. O codificador processa a entrada, e o decodificador gera a saída correspondente, permitindo transformações complexas de texto.

Crie um ranking das organizações que mais apareceram na seção "Mercado" no primeiro trimestre de 2015.

```python
!pip install transformers
!pip install kaggle

import pandas as pd
import csv
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline
import collections
```

```python
# Baixar e extrair os dados
!kaggle datasets download -d marlesson/news-of-the-site-folhauol
!unzip news-of-the-site-folhauol.zip

# Carregando dados
df = pd.read_csv('articles.csv', encoding='utf-8')
```

```python
# Filtrando os dados
df_filtered = df[(df['date'] >= '2015-01-01') & (df['date'] < '2015-04-01') & (df['category'] == 'mercado')]
```

```python
# Carregando o modelo e o tokenizer
model_name = "monilouise/ner_pt_br"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForTokenClassification.from_pretrained(model_name)

# Pipeline NER
ner_pipeline = pipeline("ner", model=model, tokenizer=tokenizer, grouped_entities=True)
```

```python
# Função para extrair organizações
def extract_organizations(text):
    entities = ner_pipeline(text)
    organizations = [entity['word'] for entity in entities if entity['entity_group'] == 'ORG']
    return organizations
```

```python
# Extrair e armazenar as organizações
df_filtered['organizations'] = df_filtered['text'].apply(extract_organizations)
```

```python
df_filtered.to_csv("organizations.csv", index=False)
```

```python
# Contando a frequência das organizações
def count_word_frequency(data):
    counter = collections.Counter()
    for organizations in df_filtered['organizations']:
        for organization in organizations:
            counter[organization] += 1
    return counter

# Contanto quantidade de organizações
quant_org = count_word_frequency(df)

# Convertendo para DataFrame e ordenando pelo ranking
rankk_org = pd.DataFrame(quant_org.items(), columns=['Word', 'Frequency'])
rankk_org = rankk_org.sort_values(by='Frequency', ascending=False).reset_index(drop=True)
```
